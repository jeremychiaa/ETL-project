```python
# Dependencies
import requests
import json
from config import gkey
import pandas as pd
from pprint import pprint
```

```python
# Read into csv file
crash_file = "Resources/Crashes_Last_Five_Years.csv"
crash_df = pd.read_csv(crash_file)
crash_df.head()
```

```python
# Using boolean mask to find date range
crash_df["ACCIDENT_DATE"] = pd.to_datetime(crash_df["ACCIDENT_DATE"]) 

# Set date within 2018
start_date = "1/1/2018"
end_date = "31/12/2018"
mask = (crash_df["ACCIDENT_DATE"] > start_date) & (crash_df["ACCIDENT_DATE"] <= end_date)

crash_df = crash_df.loc[mask]
crash_df = crash_df.reset_index(drop=True)
crash_df
```

```python
# Retrieve suburbs and postcodes from lat and lng using Google Reverse Geocoding API

for index, row in crash_df.iterrows():
    lat = row["LATITUDE"]
    lng = row["LONGITUDE"]

    query_url = f"https://maps.googleapis.com/maps/api/geocode/json?latlng={lat},{lng}&key={gkey}"

    locations = requests.get(query_url).json()
    
    try:

        if len(locations["results"][0]["formatted_address"].split(",")) == 3:
            crash_df.loc[index, "SUBURB"] = locations["results"][0]["formatted_address"].split(",")[1].split("VIC")[0]
            crash_df.loc[index, "POSTCODE"] = locations["results"][0]["formatted_address"].split(",")[1].split("VIC")[1]
        else:
            crash_df.loc[index, "SUBURB"] = locations["results"][0]["formatted_address"].split(",")[2].split("VIC")[0]
            crash_df.loc[index, "POSTCODE"] = locations["results"][0]["formatted_address"].split(",")[2].split("VIC")[1]

    except (KeyError, IndexError):
         print(f"Missing field/result... skipping index number {index}.")
```

```python
# Create a filtered dataframe from specific columns
cd_cols = ['OBJECTID',
 'ACCIDENT_NO',
 'ACCIDENT_DATE',
 'ACCIDENT_TIME',
 'ACCIDENT_TYPE',
 'DAY_OF_WEEK',
 'ROAD_GEOMETRY',
 'SEVERITY',
 'SPEED_ZONE',
 'NODE_ID',
 'LONGITUDE',
 'LATITUDE',
 'SUBURB',
 'POSTCODE',
 'NODE_TYPE',
 'LGA_NAME',
 'REGION_NAME',
 'TOTAL_PERSONS',
 'INJ_OR_FATAL',
 'FATALITY',
 'SERIOUSINJURY',
 'OTHERINJURY',
 'NONINJURED',
 'MALES',
 'FEMALES',
 'BICYCLIST',
 'PASSENGER',
 'DRIVER',
 'PEDESTRIAN',
 'PILLION',
 'MOTORIST',
 'OLD_PEDESTRIAN',
 'OLD_DRIVER',
 'YOUNG_DRIVER',
 'ALCOHOL_RELATED',
 'NO_OF_VEHICLES',
 'HEAVYVEHICLE',
 'PASSENGERVEHICLE',
 'MOTORCYCLE',
 'PUBLICVEHICLE']

cd_transformed = crash_df[cd_cols].copy()

# Clean the data by dropping duplicates and setting the index
cd_transformed.drop_duplicates("OBJECTID", inplace=True)

cd_transformed.head()
```

```python
# Check for missing values
cd_transformed.count()
```

```python
# Drop missing values from postcodes and node types
cd_transformed = cd_transformed.dropna(subset=["POSTCODE", "NODE_TYPE"])
```

```python
# Drop entire day of week column
cd_transformed = cd_transformed.dropna(axis="columns")
```

```python
# Convert column names to lowercase
cd_transformed.columns = cd_transformed.columns.str.lower()
cd_transformed
```

```python
# Export dataframe to csv
cd_transformed.to_csv("output/transformed_crash_data.csv", index=False)
```
